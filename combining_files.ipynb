## Combine in dataframes

```python
import pandas as pd
import numpy as np
import math
```

```python
#Read in seperate dfs
nfl_2021 = pd.read_csv('nfl_scrape.csv')
nfl_2015 = pd.read_csv('nfl_scrape_2015_2020.csv')
nfl_2014 = pd.read_csv('nfl_scrape_2014_2025.csv')
nfl_2010 = pd.read_csv('nfl_scrape_test.csv')

dfs = [nfl_2010, nfl_2014, nfl_2015, nfl_2021]
nfl = nfl_2010.copy()

#Append dfs together
for i in range(len(dfs)-1):
    nfl = nfl.append(dfs[i+1])
```

```python
#Sort and save csv
nfl = nfl.sort_values(by='year')
nfl.to_csv('nfl_data_full.csv')
```

## Add missing lines

```python
spreads = pd.read_csv('spreadspoke_scores.csv')
spreads.head()
```

```python
# Reduce spreads to years of interest
spreads_filtered = spreads[spreads['schedule_season']>2009]
spreads_filtered = spreads_filtered[spreads_filtered['schedule_season']<2021]
spreads_filtered = spreads_filtered[spreads_filtered['schedule_playoff']==False]
spreads_filtered.shape
```

```python
team_list = np.sort(spreads_filtered['team_home'].unique())
print(team_list)
```

```python
abr_list = np.sort(nfl['home_team'].unique())
abr_list
```

```python
#Create dictionary mapping team names to abreviation
team_dict = {'Arizona Cardinals': 'ARI', 'Atlanta Falcons': 'ATL', 'Baltimore Ravens': 'BAL', 'Buffalo Bills': 'BUF',
 'Carolina Panthers': 'CAR', 'Chicago Bears': 'CHI', 'Cincinnati Bengals': 'CIN',
 'Cleveland Browns': 'CLE', 'Dallas Cowboys': 'DAL', 'Denver Broncos': 'DEN', 'Detroit Lions': 'DET',
 'Green Bay Packers': 'GB', 'Houston Texans': 'HOU', 'Indianapolis Colts': 'IND',
 'Jacksonville Jaguars': 'JAX', 'Kansas City Chiefs': 'KC', 'Las Vegas Raiders': 'LV',
 'Los Angeles Chargers': 'LAC', 'Los Angeles Rams': 'LAR', 'Miami Dolphins': 'MIA',
 'Minnesota Vikings': 'MIN', 'New England Patriots': 'NE', 'New Orleans Saints': 'NO',
 'New York Giants': 'NYG', 'New York Jets': 'NYJ', 'Oakland Raiders': 'OAK', 'Philadelphia Eagles': 'PHI',
 'Pittsburgh Steelers': 'PIT', 'San Diego Chargers': 'SD', 'San Francisco 49ers': 'SF',
 'Seattle Seahawks': 'SEA', 'St. Louis Rams': 'STL', 'Tampa Bay Buccaneers': 'TB',
 'Tennessee Titans': 'TEN', 'Washington Football Team': 'WSH', 'Washington Redskins': 'WSH'}
```

```python
# Replacing team names with abreviations
for i in range(spreads_filtered.shape[0]):
    spreads_filtered['team_home'].iloc[i] = team_dict[spreads_filtered['team_home'].iloc[i]]
    spreads_filtered['team_away'].iloc[i] = team_dict[spreads_filtered['team_away'].iloc[i]]
```

```python
spreads_filtered.head()
```

```python
# Add column to match based off week, home team and away team
spreads_filtered['match'] = None
spreads_filtered['match'] = spreads_filtered['schedule_week'] + spreads_filtered['team_home'] + spreads_filtered['team_away']
spreads_filtered.head()
```

```python
# Add same matching column to nfl data
nfl['match'] = None
for k in range(nfl.shape[0]):
    nfl['match'].iloc[k] = str(nfl['week'].iloc[k]).replace(" ", "") + str(nfl['home_team'].iloc[k]).replace(" ", "") + str(nfl['away_team'].iloc[k]).replace(" ", "")
nfl.head()
```

```python
# Filter spreads to only the columns that matter
spreads_df = spreads_filtered[['team_favorite_id', 'spread_favorite', 'over_under_line', 'match']]
spreads_df.head()
```

```python
#Combine dataframes based off matching column
nfl_test = nfl.join(spreads_df.set_index('match'), on = 'match')
nfl_test.head()
```

```python
# Fill in values in the full data frame so there is only one column for lines
for t in range(nfl_test.shape[0]):
    if math.isnan(nfl_test['favorite'].iloc[t]):
        nfl_test['favorite'].iloc[t] = nfl_test['team_favorite_id'].iloc[t]
        nfl_test['line'].iloc[t] = nfl_test['spread_favorite'].iloc[t]
        nfl_test['over/under'].iloc[t] = nfl_test['over_under_line'].iloc[t]
nfl_test.head()
```

```python
# Drop unecessary columns
nfl_test = nfl_test.drop(['match', 'team_favorite_id', 'spread_favorite', 'over_under_line'], axis = 1)
nfl_test.head()
```

```python
# Save results to dataframe
nfl_test.to_csv('nfl_data_full.csv')
```
