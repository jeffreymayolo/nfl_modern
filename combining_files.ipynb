### Data cleaning

```python
nfl_test = pd.read_csv('nfl_data_full.csv')
nfl_test.shape
```

```python
nfl_test.head()
```

```python
#Dropping NA values
nfl_clean = nfl_test.dropna()
```

```python
nfl_clean.shape
```

```python
#Converting strings to ints
str_to_int = ['winning_score', 'losing_score', 'line', 'over/under']
for col in str_to_int:
    nfl_clean[col] = nfl_clean[col].astype(float)
```

```python
# Add variable for if the over hit
nfl_clean['over'] = None
nfl_clean['total_score'] = None
for j in range(nfl_clean.shape[0]):
    nfl_clean['total_score'].iloc[j] = nfl_clean['winning_score'].iloc[j]+nfl_clean['losing_score'].iloc[j]
    if nfl_clean['total_score'].iloc[j] > nfl_clean['over/under'].iloc[j]:
        nfl_clean['over'].iloc[j] = 1
    elif nfl_clean['total_score'].iloc[j] == nfl_clean['over/under'].iloc[j]:
        nfl_clean['over'].iloc[j] = np.nan
    else:
        nfl_clean['over'].iloc[j] = 0
```

```python
old_dict = {'OAK': 'LV', 'STL': 'LAR', 'SD': 'LAC'}
for i in range(nfl_clean.shape[0]):
    if nfl_clean['winning_team'].iloc[i] in list(old_dict.keys()):
        nfl_clean['winning_team'].iloc[i] = old_dict[nfl_clean['winning_team'].iloc[i]]
    if nfl_clean['losing_team'].iloc[i] in list(old_dict.keys()):
        nfl_clean['losing_team'].iloc[i] = old_dict[nfl_clean['losing_team'].iloc[i]]
```

```python
if 'Unnamed: 0' in nfl_clean.columns:
    nfl_clean = nfl_clean.drop(['Unnamed: 0'], axis = 1)
nfl_clean.head()
```

```python
# Add Variable for score differential and spread

nfl_clean['fav_margin'] = None
nfl_clean['fav_cover'] = None
for i in range(nfl_clean.shape[0]):
    if nfl_clean['winning_team'].iloc[i]==nfl_clean['favorite'].iloc[i]:
        nfl_clean['fav_margin'].iloc[i] = nfl_clean['winning_score'].iloc[i] - nfl_clean['losing_score'].iloc[i]
    else:
        nfl_clean['fav_margin'].iloc[i] = nfl_clean['losing_score'].iloc[i] - nfl_clean['winning_score'].iloc[i]
    if nfl_clean['fav_margin'].iloc[i] > abs(nfl_clean['line'].iloc[i]):
        nfl_clean['fav_cover'].iloc[i] = 1
    elif nfl_clean['fav_margin'].iloc[i] == abs(nfl_clean['line'].iloc[i]):
        nfl_clean['fav_cover'].iloc[i] = np.nan
    else:
        nfl_clean['fav_cover'].iloc[i] = 0
```

```python
nfl_clean.head(-20)
```

```python
nfl_clean = nfl_clean.drop(['game_id'], axis = 1)
for i in range(nfl_clean.shape[1]):
    print(nfl_clean.columns[i] + ': ' + str(type(nfl_clean.iloc[0][i])))
```

```python
effs = ['home_3rd down efficiency', 'away_3rd down efficiency', 'home_4th down efficiency', 'away_4th down efficiency', 'home_Comp-Att', 'away_Comp-Att', 'home_Red Zone (Made-Att)', 'away_Red Zone (Made-Att)']

for col in effs:
    for i in range(nfl_clean.shape[0]):
        eff = nfl_clean[col].iloc[i].split('-')
        try:
            nfl_clean[col].iloc[i] = float(eff[0])/float(eff[1])
        except Exception as e:
            nfl_clean[col].iloc[i] = np.nan
```

```python
varis = ['home_Sacks-Yards Lost', 'away_Sacks-Yards Lost', 'home_Penalties', 'away_Penalties']
for col in varis:
    for i in range(nfl_clean.shape[0]):
        vari = nfl_clean[col].iloc[i].split('-')
        nfl_clean[col].iloc[i] = int(vari[0])
```

```python
for i in range(nfl_clean.shape[0]):
    poss = nfl_clean['home_Possession'].iloc[i].split(':')
    nfl_clean['home_Possession'].iloc[i] = float(poss[0]) + float(poss[1])/60
    poss = nfl_clean['away_Possession'].iloc[i].split(':')
    nfl_clean['away_Possession'].iloc[i] = float(poss[0]) + float(poss[1])/60
```
